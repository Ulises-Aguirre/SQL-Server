## CONSULTAS EN SQL SERVER

### Indicar cuántos cursos y carreras tiene el área de Data. 
- Renombrar la nueva columna como cant_asignaturas.

```python
# Crear la cadena de conexión usando SQLAlchemy
connection_string = 'mssql+pyodbc://LAPTOP-S450FOBN/CoderHouse?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server'
# Crear el motor de conexión
engine = create_engine(connection_string)
# Consulta SQL
sql_query = """
SELECT A.Tipo, COUNT(A.Tipo) AS Cant_asignaturas
FROM Asignaturas AS A
INNER JOIN Area AS AR ON AR.AreaID = A.Area
WHERE AR.Nombre = 'Data'
GROUP BY A.Tipo
UNION ALL
SELECT 'Total', COUNT(A.Tipo) AS Cant_asignaturas
FROM Asignaturas AS A
INNER JOIN Area AS AR ON AR.AreaID = A.Area
WHERE AR.Nombre = 'Data';
"""
# Leer los datos en un DataFrame
consulta = pd.read_sql_query(sql_query, engine)
# Mostrar el DataFrame
consulta
```

### Se requiere saber cual es el nombre, el documento de identidad y el teléfono de los estudiantes que son profesionales en agronomía y que nacieron entre el año 1970 y el año 2000.0

```python
# Crear la cadena de conexión usando SQLAlchemy
connection_string = 'mssql+pyodbc://LAPTOP-S450FOBN/CoderHouse?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server'
# Crear el motor de conexión
engine = create_engine(connection_string)
# Consulta SQL
sql_query = """
SELECT E.Nombre, E.Documento, E.Telefono
FROM Estudiantes AS E
INNER JOIN Profesiones AS P ON P.ProfesionesID = E.Profesion
WHERE E.[Fecha de Nacimiento] BETWEEN '1970-01-01' AND '2000-12-31' 
AND P.Profesiones = 'Agronomo Agronoma';
"""
# Leer los datos en un DataFrame
consulta = pd.read_sql_query(sql_query, engine)
# Mostrar el DataFrame
consulta
```

### Se requiere un listado de los docentes que ingresaron en el año 2021 y concatenar los campos nombre y apellido. 
- El resultado debe utilizar un separador: guión (-). Ejemplo: Elba-Jimenez. 
- Renombrar la nueva columna como Nombres_Apellidos. 
- Los resultados de la nueva columna deben estar en mayúsculas.

```python
# Crear la cadena de conexión usando SQLAlchemy
connection_string = 'mssql+pyodbc://LAPTOP-S450FOBN/CoderHouse?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server'
# Crear el motor de conexión
engine = create_engine(connection_string)
# Consulta SQL
sql_query = """
SELECT UPPER(CONCAT(Nombre, '-', Apellido)) AS Nombre_Apellidos
FROM Staff
WHERE YEAR([Fecha Ingreso]) = 2021;
"""
# Leer los datos en un DataFrame
consulta = pd.read_sql_query(sql_query, engine)
# Mostrar el DataFrame
consulta
```

### Indicar la cantidad de encargados de docentes y de tutores. 
- Renombrar la columna como CantEncargados.  
- Quitar la palabra ”Encargado ”en cada uno de los registros. 

```python
# Crear la cadena de conexión usando SQLAlchemy
connection_string = 'mssql+pyodbc://LAPTOP-S450FOBN/CoderHouse?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server'
# Crear el motor de conexión
engine = create_engine(connection_string)
# Consulta SQL
sql_query = """
SELECT REPLACE(Tipo, 'Encargado', '')AS Tipo, COUNT(Nombre) AS CantEncargados FROM Encargado
GROUP BY Tipo
"""
# Leer los datos en un DataFrame
consulta = pd.read_sql_query(sql_query, engine)
# Mostrar el DataFrame
consulta
```

### Indicar cuál es el precio promedio de las carreras y los cursos por jornada. 
- Renombrar la nueva columna como Promedio. 
- Ordenar los promedios de Mayor a menor. 

```python
# Crear la cadena de conexión usando SQLAlchemy
connection_string = 'mssql+pyodbc://LAPTOP-S450FOBN/CoderHouse?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server'
# Crear el motor de conexión
engine = create_engine(connection_string)
# Consulta SQL
sql_query = """
SELECT Tipo, Jornada, ROUND(AVG(Costo),2) AS Costo_Promedio FROM Asignaturas
GROUP BY Tipo, Jornada
ORDER BY Costo_Promedio DESC
"""
# Leer los datos en un DataFrame
consulta = pd.read_sql_query(sql_query, engine)
# Mostrar el DataFrame
consulta
```

### Se requiere calcular la edad de los estudiantes en una nueva columna. 
- Renombrar a la nueva columna Edad. 
- Filtrar solo los que son mayores de 18 años. 
- Ordenar de Menor a Mayor.

```python
# Crear la cadena de conexión usando SQLAlchemy
connection_string = 'mssql+pyodbc://LAPTOP-S450FOBN/CoderHouse?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server'
# Crear el motor de conexión
engine = create_engine(connection_string)
# Consulta SQL
sql_query = """
SELECT *, DATEDIFF(year, [Fecha de Nacimiento], GETDATE()) AS Edad 
FROM Estudiantes
WHERE DATEDIFF(year, [Fecha de Nacimiento], GETDATE()) > 18
ORDER BY Edad DESC;
"""
# Leer los datos en un DataFrame
consulta = pd.read_sql_query(sql_query, engine)
# Mostrar el DataFrame
consulta
```

### Se desea calcular el 25% de aumento para las asignaturas del área de marketing de la jornada mañana se deben traer todos los campos, mas el de los cálculos correspondientes el porcentaje y el Nuevo costo debe estar en decimal con 3 digitos. 
- Renombrar el calculo del porcentaje con el nombre porcentaje y la suma del costo mas el porcentaje por NuevoCosto. 

```python
# Crear la cadena de conexión usando SQLAlchemy
connection_string = 'mssql+pyodbc://LAPTOP-S450FOBN/CoderHouse?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server'
# Crear el motor de conexión
engine = create_engine(connection_string)
# Consulta SQL
sql_query = """
SELECT A.Nombre, ASIG.Jornada, ASIG.Costo,
    CASE
        WHEN A.Nombre = 'Marketing Digital' THEN ASIG.Costo * 1.25
        ELSE ASIG.Costo
    END AS Nuevo_Costo
FROM Asignaturas AS ASIG
INNER JOIN Area AS A ON A.AreaID = ASIG.Area;
"""
# Leer los datos en un DataFrame
consulta = pd.read_sql_query(sql_query, engine)
# Mostrar el DataFrame
consulta
```
